```python
import pandas as pd
import numpy as np
```

```python
path = r"C:\Users\Alvin\Desktop\UT\fall2019\research project\depression\health_mdd_prac.csv"
df = pd.read_csv(path)
df
```

```python
for i in list(df.columns):
    if i != "label":
        name=i+"_delta"
        df[name]=float(0)
        for j in range(1, len(df)):
            df[name][j]=float(df[i][j]) - float(df[i][j-1])
```

```python
# convert categorical label to numerical, healthy = 1, depressed = 0
for i in range(0, len(df["label"])):
    if df["label"][i] == "healthy":
        df["label"][i] = int(1)
    else:
        df["label"][i] = int(0)
df["label"] = df["label"].astype('int32')
df
```

```python
x = df[df.columns[df.columns!="label"]].values
y = df["label"].values
print(x.shape)
print(y.shape)
```

```python
# get balanced dataset

x = x[:30000]
y = y[:30000]
```

```python
# scaling or standardization data

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
scaler = MinMaxScaler()
std = StandardScaler()
x_sca = std.fit_transform(x)
```

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, roc_auc_score, confusion_matrix

loo = LeaveOneOut()
loocv = loo.get_n_splits(x)
```

```python
# logistic regression with ridge regularization
clf = LogisticRegression(C=1.0, penalty='l2', random_state=0)
pred = cross_val_predict(clf, x_sca, y, cv=10)
clf.fit(x_sca, y)
# scores = cross_validate(clf, x_trial, y_trial, cv=10, scoring=('roc_auc','accuracy','precision','recall','f1'))
```

```python
# get metrics 

def get_performance_metrics(y_trial, pred, threshold=0.5):
    metric_names = ['AUC','Accuracy','Precision','Recall','f1-score']
    metric_values_cv = [roc_auc_score(y_trial, pred),
                    accuracy_score(y_trial, pred>threshold),
                    precision_score(y_trial, pred>threshold),
                    recall_score(y_trial, pred>threshold),
                    f1_score(y_trial, pred>threshold)]

    all_metrics = pd.DataFrame({'metrics':metric_names,
                                'cv_test_scores':metric_values_cv}, 
                               columns=['metrics','cv_test_scores']).set_index('metrics')
    print(all_metrics)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(y_trial, pred):
    roc_auc_test = roc_auc_score(y_trial, pred)
    fpr_test, tpr_test, _ = roc_curve(y_trial, pred)
    plt.figure()
    lw = 2
    plt.plot(fpr_test, tpr_test, color='darkorange',
             lw=lw, label='ROC Test (AUC = %0.4f)' % roc_auc_test)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
# get feature importance

def coef_importance(x, clf):
    selected_features = list(df.columns[df.columns!="label"])
    df_coeffs = pd.DataFrame(list(zip(selected_features, clf.coef_.flatten()))).sort_values(by=[1], ascending=False)
    df_coeffs.columns = ['feature', 'coeff']
    ax = df_coeffs.plot.barh(figsize=(6,6))
    t = np.arange(x.shape[1])
    ax.set_yticks(t)
    ax.set_yticklabels(df_coeffs['feature'])
    plt.show()
```

```python
# get confusion matrix plot
from sklearn.utils.multiclass import unique_labels

def plot_confusion_matrix(y_true, y_pred,
                          cmap=plt.cm.Blues):
    classes = np.array(["depressed", "healthy"])
    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=30, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '2d'
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="bottom",
                    color="black")
    fig.tight_layout()
    plt.show
```

```python
get_performance_metrics(y,pred)
plot_roc_curve(y,pred)
plot_confusion_matrix(y,pred)
coef_importance(x, clf)
```
